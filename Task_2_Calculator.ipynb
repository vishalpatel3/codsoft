# calculator

```python
from tkinter import *

first_number=second_number=operator=None

def get_digit(digit):
    current=l['text']
    new=current + str(digit)
    l.config(text=new)

def clear():
    l.config(text="")
    
def get_operator(op):
    global first_number,operator
    
    first_number=int(l['text'])
    operator=op
    l.config(text='')
    
def get_result():
    global first_number,second_number,operator
    second_number=int(l['text'])

    if operator=='+':
        l.config(text=str(first_number+second_number))
    elif operator=='-':
        l.config(text=str(first_number-second_number))
    elif operator=='*':
        l.config(text=str(first_number*second_number))
    else:
        if second_number==0:
            l.config(text='Error')
        else:
            l.config(text=str(round(first_number/second_number,2)))
    
root=Tk()
root.geometry('280x382')
root.title('Calculator')
root.resizable(0,0)
root.configure(bg='green')

l=Label(root,text="",bg='green',fg='white')
l.grid(row=0,column=0,columnspan=1000,pady=(50,25),sticky='e')
l.config(font=('verdana',30,'bold'))

b_7=Button(root,text='7',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(7))
b_7.grid(row=1,column=0)
b_7.config(font=('verdana',14))

b_8=Button(root,text='8',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(8))
b_8.grid(row=1,column=1)
b_8.config(font=('verdana',14))

b_9=Button(root,text='9',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(9))
b_9.grid(row=1,column=2)
b_9.config(font=('verdana',14))

b_add=Button(root,text='+',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_operator('+'))
b_add.grid(row=1,column=3)
b_add.config(font=('verdana',14))

b_4=Button(root,text='4',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(4))
b_4.grid(row=2,column=0)
b_4.config(font=('verdana',14))

b_5=Button(root,text='5',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(5))
b_5.grid(row=2,column=1)
b_5.config(font=('verdana',14))

b_6=Button(root,text='6',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(6))
b_6.grid(row=2,column=2)
b_6.config(font=('verdana',14))

b_sub=Button(root,text='-',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_operator('-'))
b_sub.grid(row=2,column=3)
b_sub.config(font=('verdana',14))

b_1=Button(root,text='1',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(1))
b_1.grid(row=3,column=0)
b_1.config(font=('verdana',14))

b_2=Button(root,text='2',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(2))
b_2.grid(row=3,column=1)
b_2.config(font=('verdana',14))

b_3=Button(root,text='3',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(3))
b_3.grid(row=3,column=2)
b_3.config(font=('verdana',14))

b_mul=Button(root,text='*',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_operator('*'))
b_mul.grid(row=3,column=3)
b_mul.config(font=('verdana',14))

b_clr=Button(root,text='C',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :clear())
b_clr.grid(row=4,column=0)
b_clr.config(font=('verdana',14))

b_0=Button(root,text='0',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_digit(0))
b_0.grid(row=4,column=1)
b_0.config(font=('verdana',14))

b_eq=Button(root,text='=',bg='red',fg='white',width=5,height=2,relief='sunken',command=get_result)
b_eq.grid(row=4,column=2)
b_eq.config(font=('verdana',14))

b_div=Button(root,text='/',bg='red',fg='white',width=5,height=2,relief='sunken',command=lambda :get_operator('/'))
b_div.grid(row=4,column=3)
b_div.config(font=('verdana',14))

root.mainloop()
```
